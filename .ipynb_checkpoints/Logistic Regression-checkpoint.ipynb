```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
from scipy.stats import logistic

np.random.seed(123)
```

```python
#y = (np.random.rand(100) < 0.3).astype(int)
```

```python
x = np.random.normal(size = (100, 1))
```

```python
def _sample(num):
    if num > -0.5 :
        return np.random.rand() < 0.7
    else:
        return np.random.rand() < 0.3

_vsample = np.vectorize(_sample)
```

```python
y = _vsample(x) * 1
```

```python
#x[y == 1] = x[y == 1] + 1.5
```

```python
plt.scatter(x, y)
plt.ylabel("outcome")
plt.xlabel("input")
plt.xlim((-6, 6))
plt.title("Outcome ~ Input")
```

Consider the following data above. There is an outcome category 1/0 that you'd like to predict based on some input data. Using some statistical models we can try to predict the probability of outcome = 1, P(outcome = 1).

In linear regression you'd try to identify a linear relationship between two continuous variables in order to model the relationship the two variables have with each other. This usually results in a line.

In the situation above, you could try to fit a linear regression model, but you may encounter some issues.

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
model = LinearRegression().fit(x, y)
```

```python
a = np.linspace(-6, 6)
b = model.predict(a.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.plot(a, b)
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

Some of the problems this method has is that the predicted values are not bounded by (0,1). If we're modeling probabilities, predicted probability values less than zero or greater than one do not make sense. Also, truncating these values would not make sense, since a value of 1 or 0 would mean complete certainty (which we do not have).

Therefore, we need to find a statistical model that is naturally bounded between (0,1). Logistic regression provides a solution since it uses the logistic (or sigmoid) distribution to estimate probability values between (0,1)

```python
plt.plot(a, logistic.cdf(a), '--')
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

```python
logistic_model = LogisticRegression().fit(x, y)
```

```python
a = np.linspace(-6, 6)
b = logistic_model.predict(a.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.plot(a, b)
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

Suppose you have a random variable Y which takes values (0,1) with probability of Y = 1, $P(Y = 1) = p_i$. Assume this response is conditional on some $k$ predictors $x_1, ..., x_k$. Using a method analogous to linear regression, we can model $p_i = w_1 x_1 + ... + w_k x_k$. However, unlike linear regression, we require that our outcome variable $p_i$ be bounded between (0, 1).
