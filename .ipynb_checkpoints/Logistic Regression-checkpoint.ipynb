```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
from scipy.stats import logistic

np.random.seed(123)
```

```python
#y = (np.random.rand(100) < 0.3).astype(int)
```

```python
x = np.random.normal(size = (100, 1))
#x = np.linspace(-4, 4).reshape(-1, 1)
```

```python
def _sample(num):
    if num > -0.5 :
        return np.random.rand() < 0.6
    else:
        return np.random.rand() < 0.3

_vsample = np.vectorize(_sample)
```

```python
y = _vsample(x) * 1
```

```python
#x[y == 1] = x[y == 1] + 1.5
```

```python
plt.scatter(x, y)
plt.ylabel("outcome")
plt.xlabel("input")
plt.xlim((-6, 6))
plt.title("Outcome ~ Input")
```

Consider the following data above. There is an outcome category 1/0 that you'd like to predict based on some input data. Using some statistical models we can try to predict the probability of outcome = 1, P(outcome = 1).

In linear regression you'd try to identify a linear relationship between two continuous variables in order to model the relationship the two variables have with each other. This usually results in a line. The relationship usually looks like 

$$
y = w_1 x_1 + ... + w_k x_k
$$

for k predictors.

In the situation above, you could try to fit a linear regression model, but you may encounter some issues.

$$
p = w_1 x_1 + ... + w_k x_k
$$

where $p = Prob(outcome = 1)$

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
model = LinearRegression().fit(x, y)
```

```python
a = np.linspace(-6, 6)
b = model.predict(a.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.plot(a, b)
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

Some of the problems this method has is that the predicted values are not bounded by (0,1). If we're modeling probabilities, predicted probability values less than zero or greater than one do not make sense. Also, truncating these values would not make sense, since a value of 1 or 0 would mean complete certainty (which we do not have).

Therefore, we need to find a statistical model that is naturally bounded between (0,1). Logistic regression provides a solution since it uses the logistic (or sigmoid) distribution to estimate probability values between (0,1)

```python
plt.plot(a, logistic.cdf(a), '--')
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

```python
logistic_model = LogisticRegression().fit(x, y)
```

```python
a = np.linspace(-6, 6)
b = logistic_model.predict(a.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.plot(a, b)
plt.ylabel("P(outcome = 1)")
plt.xlabel("input")
plt.title("P(Outcome = 1) ~ Input")
```

Suppose you have a random variable $Y$ which takes values (0,1) with probability of Y = 1, $P(Y = 1|X) = p$. Assume this response is conditional on some $k$ predictors $x_1, ..., x_k$. Using a method analogous to linear regression, we can model $p = w_1 x_1 + ... + w_k x_k$. However, unlike linear regression, we require that our outcome variable $p$ be bounded between (0, 1).

What we can do instead of modeling $p$ directly is use a *link function* called the *logit* (or log-odds)

$$
z = g(p) = log \left(\frac{p}{1-p} \right) = w_1 x_1 + ... + w_k x_k
$$

Then, solving for $p$ we obtain 

$$
p = g^{-1}(z) = \frac{e^z}{1 + e^z}
$$

which is the *logistic function*, and thus we obtain logistic regression.

The benefits of having the logistic function in this model are that as the values of z approach $\infty$ (or $- \infty$) the value of $p$ approaches 1 (or 0) but never equals 1 (or 0). With logistic regression would not have an issue of $p \geq 1$, or saying that $Y = 1$ with absolute certainty.

Logistic regression is part of a family of statistical models called *generalized linear models* (GLMs) because it expresses the explanatory (input) variables as a linear function.

$$
g(p) = w_1 x_1 + ... + w_k x_k
$$

The differences between linear and logistic regression are the link functions and method of parameter estimations. Linear regression uses the identity link function, and logistic regression uses the logit link function. Linear regression estimates parameters by minimizing the residuals through a method called Ordinary Least Squares (OLS). Residuals don't necessary fit in the logistic regression methodology, so parameters are instead estimated via maximum likelihood (ML).

Applications of the two models also differ. Linear regression assumes a continuous random variable $Y$. Logistic regression assumes a dichotomous random variable, such that $Y$ has a bernoulli distirubition (or where $Y$ has a binomial distribution if $Y$ is a proportion).
